```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sales data
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000, 350000, 480000, 520000, 650000],  
    [2, 120000, 140000, 160000, 190000, 220000, 340000, 460000, 590000],  
    [3, 200000, 230000, 260000, 300000, 420000, 540000, 560000, 690000],  
    [4, 180000, 210000, 240000, 270000, 520000, 440000, 360000, 590000],  
    [5, 160000, 185000, 205000, 230000, 420000, 340000, 360000, 790000],  
])

# Separate IDs and sales
ids = sales_data[:, 0]
sales = sales_data[:, 1:]

# Create x-axis for time points
time_points = np.arange(1, sales.shape[1] + 1)

# Plot each ID's sales over time
plt.figure(figsize=(10, 6))
for i in range(sales.shape[0]):
    plt.plot(time_points, sales[i], marker='o', label=f'ID {int(ids[i])}')

plt.title('Sales Over Time by ID')
plt.xlabel('Time Point')
plt.ylabel('Sales Amount')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


```
